```
"""
Test PaliGemma language capabilities to talk using a Instance from PI05Policy
"""
import torch
from transformers import AutoTokenizer
from lerobot.policies.pi05.modeling_pi05 import PI05Policy
```

```
policy = PI05Policy.from_pretrained("lerobot/pi05_base")

device = policy.config.device
paligemma = policy.model.paligemma_with_expert.paligemma.to(device).eval()

# Use the same tokenizer as PI05
tokenizer = AutoTokenizer.from_pretrained("google/paligemma-3b-pt-224")
```

```
prompt = "Hola como te llamas? "
inputs = tokenizer(prompt, return_tensors="pt").to(device)

print(f"\nPrompt: {prompt}")
print("Generating...")

with torch.no_grad():
    generated_ids = paligemma.generate(
        **inputs,
        max_new_tokens=400,
        do_sample=True,
        temperature=0.1,
    )

text = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
print(f"\nGenerated text:\n{text}")
```
